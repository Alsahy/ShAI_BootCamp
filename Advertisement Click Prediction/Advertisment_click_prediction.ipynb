# Logistic Regression

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

** Split the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
```

```python
X = ad_data.drop('Clicked on Ad', axis=1)
y = ad_data['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.drop(['Country', 'Ad Topic Line', 'City', 'Timestamp'], axis=1, inplace = True)
X_test.drop(['Country', 'Ad Topic Line', 'City', 'Timestamp'], axis=1, inplace = True)
```

feature scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

** Train and fit a logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

logistic regression hyperparameters tuning

https://machinelearningmastery.com/hyperparameters-for-classification-machine-learning-algorithms/

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedStratifiedKFold
```

```python
solvers = ['newton-cg', 'lbfgs', 'liblinear']
penalty = ['l2']
c_values = [100, 10, 1.0, 0.1, 0.01]
```

```python
grid = dict(solver=solvers,penalty=penalty,C=c_values)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
grid_search = GridSearchCV(estimator=logmodel, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy',error_score=0)
grid_result = grid_search.fit(X_train, y_train)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
y_pred = grid_result.predict(X_test)
```

** Create a Confusion Matrix for the model.**

```python
conf = confusion_matrix(y_test, y_pred)
recall = conf[1,1]/(conf[1,1]+conf[1,0])
print('Recall: ', recall)
precision = conf[1,1]/(conf[1,1]+conf[0,1])
print('Precision: ', precision)
f1 = 2*(precision*recall)/(precision+recall)
print('F1: ', f1)
```

```python
f1_KNN = 2*(precision_KNN*recall_KNN)/(precision_KNN+recall_KNN)
print('F1: ', f1_KNN)
```
